```python
# 1
data_list = pd.read_csv('names.csv').values.tolist()

# 2
data_dict = {}
for data in data_list:
    if data[3] in data_dict:
        data_dict[data[3]] += data[4]
    else:
        data_dict[data[3]] = data[4]
data_dict
```

## Exercise 4

 1. Να διαβάσετε το αρχείο ΝΒΑ players από τον φάκελο Day 1.
 2. Αφού υπολογίσετε το BMI όπως κάναμε στο μάθημα να υπολογίσετε και να τυπώσετε τη μέση τιμή, τη διάμεσο και την τυπική απόκλιση του BMI

```python
df_nba = pd.read_csv('../Day 1/nba-players.csv')
df_nba['Height (cm)'] = df_nba[' Height (Inches)']*2.54
df_nba['Weight (kg)'] = df_nba[' Weight']*0.453592
df_nba[' BMI'] = df_nba['Weight (kg)']/(df_nba['Height (cm)']/100)**2
print('BMI mean \t\t= {}\nBMI median \t\t= {}\nBMI standard deviation \t= {}'\
                  .format(df_nba[' BMI'].mean(), df_nba[' BMI'].median(), df_nba[' BMI'].std()))
```
